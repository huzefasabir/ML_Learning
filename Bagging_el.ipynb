```python
import pandas as pd
import matplotlib as mat
from sklearn.datasets import make_moons
import seaborn as sb
```

```python
x,y=make_moons(n_samples=1000,noise=0.2)
```

```python
df={"x1":x[:,0],"x2":x[:,1],"y":y}
df
```

```python
ds=pd.DataFrame(df)
```

```python
sb.scatterplot(x="x1",y="x2",hue="y",data=ds)
mat.show()
```

```python
x_a=ds.iloc[:,:-1]
y_a=ds["y"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_a,y_a,train_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
bg=BaggingClassifier(estimator=SVC(),n_estimators=100)
bg.fit(x_train,y_train)
```

```python
bg.score(x_train,y_train),bg.score(x_test,y_test)
```

```python
sv=SVC()
sv.fit(x_train,y_train)
```

```python
sv.score(x_train,y_train),sv.score(x_test,y_test)
```

```python
rf=RandomForestClassifier(n_estimators=50)
rf.fit(x_train,y_train)
```

```python
rf.score(x_train,y_train),rf.score(x_test,y_test)
```

REGRESSION

```python
data=pd.DataFrame(pd.read_excel("cgpa_package.xlsx"))
data.head()
```

```python
x=data.iloc[:,:-1]
y=data["Package (LPA)"]
x,y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
```

```python
bg=BaggingRegressor(estimator=LinearRegression(),n_estimators=40)
bg.fit(x_train,y_train)
```

```python
bg.score(x_train,y_train),bg.score(x_test,y_test)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train),lr.score(x_test,y_test)
```

```python
rf=RandomForestRegressor(n_estimators=50)
rf.fit(x_train,y_train)
rf.score(x_test,y_test),rf.score(x_train,y_train)
```
