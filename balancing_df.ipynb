Balancing dataset


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mat
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score,recall_score,f1_score
```

Creating a model without balancing

```python
df=pd.DataFrame(pd.read_csv("E:\Study\AIDS\DS\ML\Social_Network_Ads.csv"))
df.head()
```

```python
df.drop(columns=["User ID","Gender"],inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df["Purchased"].value_counts()
```

```python
x=df.iloc[:,:-1]
y=df["Purchased"]
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.2,random_state=15)
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
lr.predict([[19,19000]])
```

```python
precision_score(y_test,lr.predict(x_test))
```

```python
recall_score(y_test,lr.predict(x_test))
```

```python
f1_score(y_test,lr.predict(x_test))
```

NOW Balancing the dataset

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
ru=RandomUnderSampler()
ru_x,ru_y=ru.fit_resample(x,y)
```

```python
ru_y.value_counts()
```

Training the model on balanced dataset

```python
x1_train,x1_test,y1_train,y1_test=train_test_split(ru_x,ru_y,test_size=0.2,random_state=20)
```

```python
lr1=LogisticRegression()
lr1.fit(x1_train,y1_train)
```

```python
lr1.score(x1_test,y1_test)
```

```python
precision_score(y1_test,lr1.predict(x1_test))
```

```python
recall_score(y_test,lr.predict(x_test))
```

```python
f1_score(y_test,lr.predict(x_test))
```
